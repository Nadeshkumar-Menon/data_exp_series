Replace the NaN values with 0.
As the reviews of a given film accumulate, the Rotten Tomatoes score measures the percentage that are more positive than negative, and assigns an overall fresh or rotten rating to the movie. 
Scores of over 60 percent are considered fresh, and scores of 59 percent and under are rotten.

```python
df_mod['Rotten Tomatoes']=df_mod['Rotten Tomatoes'].fillna(str(0))
```

For the Age values conatining NaN ,for simplicity assume it to be 'all'

```python
df_mod['Age']=df_mod['Age'].fillna("all")
```

For tha NaN values in IMDB rating,lets assume them to be 4.0

```python
df_mod['IMDb']=df_mod['IMDb'].fillna(4.0)
df_mod.head()
```

Seperate the data into each of the streaming services to dispay the number of shows aired..

```python
netflix_data=df_mod.loc[df_mod["Netflix"]==1]
unneeded_columns=["Hulu","Prime Video","Disney+"]
netflix_data=netflix_data.drop(unneeded_columns,axis=1)
netflix_count=netflix_data.shape[0]
```

```python
hulu_data=df_mod.loc[df_mod["Hulu"]==1]
unneeded_columns=["Netflix","Prime Video","Disney+"]
hulu_data=hulu_data.drop(unneeded_columns,axis=1)
hulu_count=hulu_data.shape[0]
```

```python
prime_data=df_mod.loc[df_mod["Prime Video"]==1]
unneeded_columns=["Hulu","Netflix","Disney+"]
prime_data=prime_data.drop(unneeded_columns,axis=1)
prime_count=prime_data.shape[0]
```

```python
disney_data=df_mod.loc[df_mod["Disney+"]==1]
unneeded_columns=["Hulu","Prime Video","Netflix"]
disney_data=disney_data.drop(unneeded_columns,axis=1)
disney_count=disney_data.shape[0]
```

```python
labels=['Netflix','Hulu','Prime Video','Disney+']
size=[netflix_count,hulu_count,prime_count,disney_count]
explode=(0,0,0.1,0)
fig1=plt.subplot()
fig1.pie(size,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
fig1.axis('equal')
plt.show()
```
